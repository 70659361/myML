```python
def addSpace(s,n=1):
        out=""
        shift=0
        for i in range(len(s)):
                #print("<%d>[%s]" %(i, s[i]))
                
                if(s[i] in ([",",".","。","《","》","?","，","？"])):
                        shift+=1
                        #print("is shifted: %d" %shift)
                else:
                        k=i+shift
                        #print("i<%d>:+shift<%d>=k<%d>:" %(i,shift,k))
                        
                        if (k+1)%n == 0:
                                out+=s[i]+" "
                        else:
                                out+=s[i]
                        #print("k:%d out:<%s>" %(k,out))
        return out
```

```python
morefiles=open("./dataset/ts.txt")
for line in morefiles.readlines():
    line_sp=""
    if line != "\n":
        #print(line)
        line_sp = addSpace(line, 1)
        #print(line_sp)
        data+=line_sp
        tx+=line
print(data)
```

```python
corpus=data.lower().split('\n')
#print(corpus)
tt=Tokenizer()
tt.fit_on_texts(corpus)
total_words=len(tt.word_index)+1
print(tt.word_index)
```

```python
input_seq=[]
for line in corpus:
        #print(line)
        token_list=tt.texts_to_sequences([line])[0]
        print(token_list)
        for i in range(1, len(token_list)):
            n_gram_seq=token_list[:i+1]
            #print(n_gram_seq)
            input_seq.append(n_gram_seq)
#print(input_seq)

max_seq_len=max(len(x) for x in input_seq)
print("max len:%d" %max_seq_len)

input_seq_pad = np.array(sequence.pad_sequences(input_seq, maxlen=max_seq_len, padding='pre'))
print(input_seq_pad)
```

```python
xs=input_seq_pad[:,:-1]
#print(xs)
#print(xs.shape)
labels=input_seq_pad[:,-1]
print(labels)
print(labels.shape)
#print(xs.shape)

ys=tf.keras.utils.to_categorical(labels, num_classes=total_words)
total_words
#ys
```

```python
model = kr.models.Sequential()
model.add(tf.keras.layers.Embedding(total_words,240,input_length=max_seq_len-1))
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(150)))
model.add(ly.Dense(total_words,activation='softmax'))
print(total_words)
adam=tf.keras.optimizers.Adam(lr=0.01)
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
logs = "./logs/myNLP"
tboard_callback = tf.keras.callbacks.TensorBoard(log_dir = logs)
history=model.fit(xs,ys,epochs=60, verbose=1, callbacks=[tboard_callback])
```

```python
model.save('./model/myNLP',overwrite=True, save_format='tf',include_optimizer=True)
```
